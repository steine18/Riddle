First I'll create a few helper functions. 
n_digit will turn a number into a code n digits long. Example: n_digit(1,4) returns '0001'.

minimum_change looks for the difference between an unentered code and the last few digits entered into the . Example: minimum_change('0000', '0001') returns '1'. This will be added to a string that should include all possible codes with the fewest number of characters.

```python
def n_digit(num, n):
    #Turns a number into a string that is n digits long
    num = str(num)
    code = '0' * (n-len(num)) + num
    return code

def minimum_change(last_digits, code):
    # Looks digits of a new code in the last four digits entered.
    diff = ''
    for digit in range(len(code), 0, -1):
        if last_digits[-digit:] == code[:digit]:
            diff = code[digit:]
            break
    if len(diff) > 0:
        return diff
    else:
        return code
    
def check_all(unique_codes, condensed_combinations):
    for code in unique_codes:
        if code not in condensed_combinations:
            return False
    return True
```

# Setting up parameters

```python
num_digits = 1
all_combinations = '' # Empty String to add all possible combinations
unique_codes = [] # Will be used to check answers later
for num in range(10**num_digits): # Creates all code combinations
    code = n_digit(num, num_digits)
    all_combinations += code
    unique_codes.append(code)
condensed_combinations = '' #Empty String to add digits to previously entered codes
```

##  codes 
This will run through each code and see if it has been attempted in a previous combination of attempts. If not it will use minimum_change to add digits that do not overlap with the previous entered code. 

```python
for position in range(len(all_combinations) - 0 if num_digits == 1 else num_digits): # Moves through each position of all possible combinations 
        code = all_combinations[position:position+num_digits]
        if code in condensed_combinations:
            print(code)# Skips code that has already been entered
            continue
        elif len(condensed_combinations) < num_digits: # Avoids a IndexError when condensed_combinations is blank
            condensed_combinations += code
        else: # Adds minimum amount of digits to include 4 digit code not previously entered
            condensed_combinations += minimum_change(condensed_combinations[-num_digits:],
                                                     code)
print(condensed_combinations)
if check_all(unique_codes, condensed_combinations): #Check that all codes are found
    print('All possible codes are found in the string')
    print(f'The fewest number of digits to press it {len(condensed_combinations)}.')
```

## Check for correctness and number of repeats

```python
from re import findall
code_count = {}
mult = []

for code in unique_codes:
    code_count[code] = len(findall(code, condensed_combinations))
    if code_count[code] > 1:
        mult.append((code))
        
print(f'There are {len(mult)} codes that appear multiple times, they are {mult}')

    

```
