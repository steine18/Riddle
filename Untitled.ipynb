```python
from math import sqrt

def d_from_origin(x_y):
    x,y = x_y
    return(sqrt(x**2 + y**2))
    
```

```python
spaces = {}
space_num = [i for i in range(9*9)]
x_count = 0
X = 'X12345678Z'
Y = 'XABCDEFGHZ'
for x in range(10):
    y_count = 0
    for y in range(10):
        space = f'{Y[y_count]}{X[x_count]}'
        spaces[space] = (x,y, d_from_origin((x,y)))
        y_count += 1
    x_count += 1
```
